```python
import matplotlib.pyplot as plt
```

```python
brands = ["One plus", "Apple", "Google", "Microsoft"]
x = [22,9,40,15]
c = ["red", "green", "blue", "yellow"]
ex = [ 0, 0.1, 0, 0 ]
plt.pie(x, labels = brands, colors = c, explode=ex, shadow=True, autopct= "%.2f", startangle=90 )
plt.savefig("pie.png",pad_inches=0.3, bbox_inches="tight")
plt.show()
```

```python
import pandas as pd
```

```python
data = pd.read_excel("expense3.xlsx")
df = pd.DataFrame(data)
print(df)
```

```python
grouped_by = df.groupby("Payment Mode")["Amount"].sum()
print(grouped_by)
```

```python
plt.pie(grouped_by.values, labels=grouped_by.index, autopct= "%.2f", shadow=True, startangle=90)
plt.show()
```

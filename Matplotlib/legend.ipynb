```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [45,67,33,67,12]
y1 = [41,60,13,66,11]
y2 = [11,30,55,60,19]
y3 = [11,40,45,90,59]
```

```python
plt.figure(figsize=[8,7])
plt.plot(x,y, label="male")
plt.plot(x,y1, label="female")
plt.plot(x,y2, label="kids")
plt.plot(x,y3, label="blank")
plt.legend()
plt.show()
```

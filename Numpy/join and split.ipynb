```python
import numpy as np
a = [30,40,50]
b = [10,20,30]
print(a+b)
```

```python
arr1=np.array([[30,40],[20,10]])
arr2=np.array([[10,20],[50,40]])
print(np.concatenate([arr1,arr2]))
print("\n")
print(np.concatenate([arr1,arr2], axis=1))
print("\n")
```

```python
print(np.hstack([arr1,arr2]))
print("\n")
print(np.vstack([arr1,arr2]))
```

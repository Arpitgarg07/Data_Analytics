```python
import pandas as pd
```

```python
dict = {"keys": ["k1", "k2", "k1", "k2"],
        "Names": ["John", "Ben", "David", "Peter"],
        "Houses": ["red", "blue", "green", "yellow"],
        "Grades": ["A", "B", "C", "D"]}

df = pd.DataFrame(dict)
print(df)
```

```python
print(pd.melt(df, id_vars=["Names"], value_vars=["Houses", "Grades"], var_name="Houses&Grades", value_name="values"))
```

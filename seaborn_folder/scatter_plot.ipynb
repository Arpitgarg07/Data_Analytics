```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = sns.load_dataset("tips")
print(data)
```

```python
sns.scatterplot(data=data, x="total_bill", y="tip", hue= "time", size="size")
plt.show()
```

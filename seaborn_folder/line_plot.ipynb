```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = {"Days":[1,2,3,4,5],
        "NOP":[50,40,60,30,44]}

# 1. Create the DataFrame
df = pd.DataFrame(data)
print(df)

# 2. Pass the DataFrame (df) to Seaborn, not the dictionary (data)
sns.lineplot(data=df, x="Days", y="NOP")
plt.show()
```
